```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_excel('/Users/mraffyzeidan/Documents/Code/Modsur/Final Data.xlsx')
```

```python
data
```

```python
df = data.copy()
```

```python
df
```

```python
df.info()
```

```python
df['age'] = df['age'].apply(lambda x: '< 68 Tahun' if x == 0 else ' >= 68 Tahun')
df['creatinine_phosphokinase'] = df['creatinine_phosphokinase'].apply(lambda x: '< 2522 mcg/L' if x == 0 else 
                                    ' 2600 - 5209 mcg/L' if x == 1   else ' 5800 - 7861 mcg/L')
df['ejection_fraction'] = df['ejection_fraction'].apply(lambda x: ' 0 - 35%' if x == 0 else ' 36 - 55%' if x == 1 else ' 56 - 80%')
df['platelets'] = df['platelets'].apply(lambda x: ' < 427000 kp/mL' if x == 0 else ' 448000 - 850000 kp/mL')
df['serum_creatinine'] = df['serum_creatinine'].apply(lambda x: ' 0 - 3.4 mg/dL' if x == 0 else ' 3.5 - 6.1 mg/dL' if x == 1 else ' 6.5 - 9.4 mg/dL')
df['serum_sodium'] = df['serum_sodium'].apply(lambda x: ' 0 - 116 mEq/L' if x == 0 else ' 117 - 126 mEq/L' if x == 1 else ' 127 - 133  mEq/L'
                                              if x == 2 else ' 134 - 140 mEq/L' if x == 3 else ' 141 - 148 mEq/L')
df['anaemia'] = df['anaemia'].apply(lambda x: 'Tidak Anaemia' if x == 0 else 'Anaemia')
df['diabetes'] = df['diabetes'].apply(lambda x: 'Tidak Diabetes' if x == 0 else 'Diabetes')
df['high_blood_pressure'] = df['high_blood_pressure'].apply(lambda x: 'Tidak Hipertensi' if x == 0 else 'Hipertensi')
df['sex'] = df['sex'].apply(lambda x: 'Perempuan' if x == 0 else 'Laki-laki')
df['smoking'] = df['smoking'].apply(lambda x: 'Tidak Merokok' if x == 0 else 'Merokok')
df['DEATH_EVENT'] = df['DEATH_EVENT'].apply(lambda x: 'Hidup' if x == 0 else 'Meninggal')
```

```python
df
```

----------------------------------------------------------------------

## Smoking Analysis

```python
smoking = df[['smoking' ,'sex' ,'DEATH_EVENT']]
```

```python
smoking['Analysis1'] = smoking['smoking'] + ' - ' + smoking['DEATH_EVENT']
smoking['Analysis2'] = smoking['sex'] + ' - ' + smoking['smoking']
smoking['Analysis3'] = smoking['sex'] + ' ' + smoking['smoking'] + ' - ' + smoking['DEATH_EVENT']
```

```python
smoking
```

```python
palette_color = sns.color_palette('pastel')


def plot_donut(ax, values, labels, title):
    ax.pie(values, labels=labels, autopct='%1.1f%%', colors=palette_color, wedgeprops=dict(width=0.3))
    ax.set_title(title, fontsize=14)
    
    centre_circle = plt.Circle((0,0),0.70,fc='white')
    fig = plt.gcf()
    fig.gca().add_artist(centre_circle)


fig, ax = plt.subplots(figsize=(10, 7))
values1 = smoking['Analysis1'].value_counts().values
labels1 = smoking['Analysis1'].value_counts().index
plot_donut(ax, values1, labels1, 'Analysis1')
plt.show()


fig, ax = plt.subplots(figsize=(10, 7))
values2 = smoking['Analysis2'].value_counts().values
labels2 = smoking['Analysis2'].value_counts().index
plot_donut(ax, values2, labels2, 'Analysis2')
plt.show()


fig, ax = plt.subplots(figsize=(10, 7))
values3 = smoking['Analysis3'].value_counts().values
labels3 = smoking['Analysis3'].value_counts().index
plot_donut(ax, values3, labels3, 'Analysis3')
plt.show()
```
